```python
# load all necessary imports and classes

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# use a for loop to iterate over a collection

#retrieve the data from excel 

travel_df = pd.read_excel(r'C:\Users\user\Desktop\Data-Science-Coursework\Week One\cities.xlsx') # load the excel file
travel_df.head(10) # display the first 10 rows of the dataframe
cities = travel_df.to_dict('records')
```

```python
# retrieve and assign first 3 cities and their polulation

x_values = [cities[0]['City'], cities[1]['City'], cities[2]['City']]
y_values = [cities[0]['Population'], cities[1]['Population'], cities[2]['Population']]

plt.bar(x_values, y_values) # Create a bar chart
plt.xlabel('City') # Add a label to the x-axis
plt.ylabel('Population') # Add a label to the y-axis
plt.title('Population of Cities') # Add a title to the chart
plt.grid(True)
plt.xticks(rotation='vertical')

plt.show()
```
